### Machine Learning

#### Linear Regression (Single Variable)

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
plt.xlabel("Area")
plt.ylabel("Price")
plt.plot(df.area, df.price)
```

```python
model = LinearRegression()
model.fit(df[['area']], df.price)
```

```python
model.predict([[3500]])
```

```python
x = 3500
```

```python
m = model.coef_
```

```python
b = model.intercept_
```

```python
y = m * x + b
y
```
